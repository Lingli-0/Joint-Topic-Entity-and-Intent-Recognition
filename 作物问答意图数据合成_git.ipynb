## 一、作物问答意图数据合成

````python
import json
from random import random

import requests


def getRequest(content):
	"""请求大模型"""
	url = "https://aip.baidubce.com/rpc/2.0/ai_custom/v1/wenxinworkshop/chat/completions_pro?access_token=" + "{{your access_token=}}"
	payload = json.dumps({
		"messages": [
			{
				"role": "user",
				"content": content
			}
		]
	})
	headers = {
		'Content-Type': 'application/json'
	}

	response = requests.request("POST", url, headers=headers, data=payload)

	print(response.text)
	return response.text


def getPrompts(crop_name, num=20):
	""" 获取prompt模板"""
	return {
		"作物分布": f"假如你是农民，请提出{num}个关于{crop_name}作物分布的问题，突出作物地区分布！例如某地区{crop_name}产区分布情况如何？以json格式返回问题列表,如[\"问题1\",\"问题2\"...]。",
		"作物品种": f"假如你是农民，请提出{num}个关于{crop_name}作物品种的问题，突出种植品种！例如某地区常用{crop_name}品种和杂交种有哪些？？以json格式返回问题列表,如[\"问题1\",\"问题2\"...]。",
		"作物生长环境与习性": f"假如你是农民，请提出{num}个关于{crop_name}作物生长环境与习性的问题，突出作物生长环境与习性！ {crop_name}种子萌发时物质转化生理基础是什么？以json格式返回问题列表,如[\"问题1\",\"问题2\"...]。",
		"作物形态特征": f"假如你是农民，请提出{num}个关于{crop_name}作物形态特征的问题，突出作物形态特征！{crop_name}根系的形态、构造及其功能怎样？以json格式返回问题列表,如[\"问题1\",\"问题2\"...]。",
		"作物常见病虫害": f"假如你是农民，请提出{num}个关于{crop_name}作物常见病虫害的问题，突出什么时期、地区、会有哪些常见病虫害，而不是如何防治病虫害！某地区{crop_name}产区为害{crop_name}的病虫害有哪些？以json格式返回问题列表,如[\"问题1\",\"问题2\"...]。",
		"栽培技术": f"假如你是农民，请提出{num * 2}个关于{crop_name}栽培技术的问题，集中于播种、除草、浇水、采收运输、田间管理，如怎样防治{crop_name}倒伏？以json格式返回问题列表,如[\"问题1\",\"问题2\"...]。"
	}
	return prompt


def dealJSONStr(str):
	"""从返回结果中提取出json字符串"""
	start1 = str.find("[")
	start2 = str.find("{")
	end1 = -1
	while str.find("]", end1 + 1) != -1:
		end1 = str.find("]", end1 + 1)
	end2 = -1
	while str.find("}", end2 + 1) != -1:
		end2 = str.find("}", end2 + 1)
	b = -1
	s = -1
	if start1 == -1 and start2 == -1:
		b, s = -1, -1
	elif start1 == -1:
		b, s = start2, end2
	elif start2 == -1:
		b, s = start1, end1
	elif start1 < start2:
		b, s = start1, end1
	else:
		b, s = start2, end2
	return str[b:s + 1]
````

````python
crops = ['小麦', '玉米', '水稻']
for crop in crops:
	prompts = getPrompts(crop)
	for intent, prompt in prompts.items():
		if intent != "栽培技术": continue
		try:
			# print(f"作物名{content[0]}\n 背景知识{content[1]} \n 意图{intents[index]}\t {title}\n")
			print(f"prompt\t: {prompt}")
			res = getRequest(prompt)
			res = dealJSONStr(res)
			print(res)
			with open(f"作物合成数据/{crop}_{intent}.json", "w", encoding='utf-8') as f:
				f.write(res)
		except Exception as e:
			print(e)
# save(f"label/samples{index:02d}.json")
````

````python
import os, json
import http.client
import re

HOST = "api.deerapi.com"


def getRequestByChatgpt(prompt):
	conn = http.client.HTTPSConnection(HOST, timeout=30)
	payload = json.dumps({
		"model": "gpt-4.1",
		"messages": [{"role": "user", "content": prompt}],
		"stream": False
	})
	headers = {
		"Authorization": f"Bearer {{api-key}}",
		"Content-Type": "application/json"
	}
	conn.request("POST", "/v1/chat/completions", payload, headers)
	res = conn.getresponse()
	body = res.read().decode("utf-8")
	conn.close()
	return body


out_dir = "病虫害知识背景"
os.makedirs(out_dir, exist_ok=True)
diseases = {
	"小麦": {
		"病害": [
			"小麦锈病",
			"小麦赤霉病",
			"小麦茎基腐病",
			"小麦白粉病",
			"小麦根腐病",
			"小麦黑穗病",
			"小麦叶枯病"
		],
		"虫害": [
			"小麦蚜虫",
			"麦蜘蛛",
			"吸浆虫"
		]
	},
	"玉米": {
		"病害": [
			"玉米南方锈病",
			"玉米大斑病",
			"玉米小斑病",
			"玉米穗腐病",
			"玉米褐斑病"
		],
		"虫害": [
			"草地贪夜蛾",
			"玉米螟",
			"粘虫",
			"棉铃虫",
			"甜菜夜蛾",
			"双斑长跗萤叶甲"
		]
	},
	"水稻":
		{
			"病害": [
				"水稻白叶枯病",
				"稻瘟病",
				"稻曲病",
				"水稻纹枯病"
			],
			"虫害": [
				"稻飞虱",
				"螟虫",
				"稻纵卷叶螟",
				"稻秆潜蝇",
				"稻叶蝉",
				"稻粉虱"
			]
		}
}


def getPrompt(crop, disease_pest_type, disease_pest):
	with open(f"病虫害知识背景//{disease_pest}.json", 'r', encoding='utf-8') as f:
		data = json.load(f)
	generated_prompt = {}
	if (disease_pest_type == "病害"):
		generated_prompt = {
			"病虫害为害症状": f"假如你是一位种植{crop}的农民，请根据以下背景知识，提出20个与{disease_pest}的病害为害症状相关的问题。"
							  f"要求问题集中在病害的可见症状、发病部位、病程变化、病情判断标准等田间识别内容，不涉及病原学或防治措施。"
							  f"以json格式返回问题列表，如 ['问题1','问题2',...]。背景知识：{data[disease_pest]['病虫害为害症状']}。",

			"病害防治方法": f"假如你是一位农民，结合以下背景知识，提出20个关于{disease_pest}的病害防治方法相关的问题。"
							f"问题应聚焦于农业防治、化学防治、生物防治、抗病品种、田间操作时机等实践层面，不涉及基础理论。"
							f"以json格式返回问题列表。背景知识：{data[disease_pest]['病害防治方法']}。",

			"病害病原特征": f"假如你是一位农民，对{crop}的{disease_pest}不太了解。请根据以下背景知识，提出20个关于病害病原特征的提问，"
							f"如‘病原菌在何处越冬？’、‘病原类型是什么？’等。以json格式返回问题列表。背景知识：{data[disease_pest]['病害病原特征']}。",

			"病害侵染循环": f"请作为农民，提出20个关于{disease_pest}病害侵染循环过程的问题，"
							f"集中在传播途径、感染阶段、再侵染特征、季节性规律等方面。以json格式返回问题列表。背景知识：{data[disease_pest]['病害侵染循环']}。",

			"病害流行规律": f"假如你是一位农民，请基于以下背景知识，提出20个与{disease_pest}病害流行规律有关的问题。"
							f"问题应围绕气候、品种、栽培密度、地势、连作等田间管理因素的影响展开。以json格式返回问题列表。背景知识：{data[disease_pest]['病害流行规律']}。"
		}
	else:
		generated_prompt = {
			"虫害形态特征": f"假如你是一位种植{crop}的农民，请根据以下背景知识，提出20个关于{disease_pest}形态特征的问题，"
							f"例如‘成虫和幼虫有何差别？’、‘如何在田间识别？’。问题应贴近识别实践。以json格式返回问题列表。背景知识：{data[disease_pest]['虫害形态特征']}。",

			"虫害生活习性": f"假如你是一位农民，给出{crop}虫害生活习性背景知识，提出20个关于{disease_pest}虫害生活习性的问题。"
							f"问题可以包括越冬方式、休眠习性、产卵位置与数量、幼虫取食对象、虫害活动时间、迁飞特性、环境适应性等方面。"
							f"以json格式返回问题列表。背景知识：{data[disease_pest]['虫害生活习性']}。",

			"病虫害为害症状": f"假如你是一位农民，请根据以下背景知识，提出20个关于{disease_pest}病虫害为害症状的问题。"
							  f"问题聚焦于田间可观察特征：被害部位、损失程度、典型病斑、虫口密度判断等。以json格式返回问题列表。背景知识：{data[disease_pest]['病虫害为害症状']}。",

			"虫害发生规律": f"假如你是一位农民，请结合以下背景知识，提出20个关于{disease_pest}虫害发生规律的问题，"
							f"包括虫害发生时间、气候影响、种植模式关系、作物生育期对应规律等。以json格式返回问题列表。背景知识：{data[disease_pest]['虫害发生规律']}。",
			"虫害防治方法": f"假如你是一位种植{crop}的农民，请根据以下背景知识，提出20个关于{disease_pest}虫害防治方法的问题。"
							f"问题应聚焦于虫害综合治理的实际措施，包括农业防治（如合理轮作、清除虫源）、物理防治（如灯光诱杀、性诱剂）、"
							f"生物防治（如天敌利用、微生物制剂）、化学防治（药剂种类、喷施时期与方法）、以及和安全间隔期等方面。"
							f"以json格式返回问题列表。背景知识：{data[disease_pest]['虫害防治方法']}。"
		}
	return generated_prompt


import json
import re


def extract_questions_from_content(content: str):
	"""
    从 GPT 返回的 content 中提取问题列表
    兼容三种情况：纯 JSON、带 ```json 包裹、格式噪声
    """
	# 提取 JSON 代码块
	match = re.search(r"```json\s*(.*?)\s*```", content, re.S)
	if match:
		json_str = match.group(1)
	else:
		start = content.find("[")
		end = content.rfind("]")
		if start != -1 and end != -1:
			json_str = content[start:end + 1]
		else:
			raise ValueError("无法在 content 中找到有效 JSON 数据")

	json_str = json_str.strip()

	try:
		questions = json.loads(json_str)
	except Exception as e:
		print("⚠️ JSON 解析失败，原始字符串片段：", json_str[:200])
		raise e

	if not isinstance(questions, list):
		raise ValueError(f"解析结果不是列表，而是 {type(questions)}")

	# 修复：过滤非字符串项并清洗序号
	clean_questions = [
		re.sub(r"^\s*\d+\.\s*", "", str(q)).strip()
		for q in questions
		if q is not None
	]

	return clean_questions


for crop in diseases:
	for disease_pest_type in diseases[crop]:
		for diseases_pest in diseases[crop][disease_pest_type]:
			prompts = getPrompt(crop, disease_pest_type, diseases_pest)
			for item in prompts:
				if (item != "虫害防治方法" and item != "病害防治方法"): continue
				prompt = prompts[item]
				raw = getRequestByChatgpt(prompt)
				print(prompt, raw)
				data = json.loads(raw)
				content = data["choices"][0]["message"]["content"]
				questions = extract_questions_from_content(content)
				with open(f"病虫害合成数据//{disease_pest_type}.txt", "a", encoding="utf-8") as f:
					for q in questions:
						f.write(f"{diseases_pest}&nbsp;{item}&nbsp;{q.strip()}\n")
				print(f"保存{diseases_pest}_{item}问题列表.txt，共", len(questions), "条问题。")
````
